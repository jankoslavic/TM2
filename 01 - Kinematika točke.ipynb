```python
import matplotlib.pyplot as plt
import sympy as sym
sym.init_printing()
```

## Zgled časovno definirane lege

```python
t = sym.symbols('t')
```

Definicija lege:

```python
s = 2 + 3*t - 0.5*(t-0.5)**3
s
```

```python
sym.plot(s, (t,0,3));
```

Hitrost:

```python
v = s.diff(t)
v
```

```python
sym.plot(v, (t,0,3));
```

Pospešek:

```python
a = s.diff(t,2) # ali: v.diff(t)
a
```

```python
sym.plot(a, (t,0,3));
```
